# 12. Classify the employees into 3 categories.
 

```python
Category 1 : employees below 18.5 bmi - underweight
Category 2 : employees with bmi in the range of 18.5 to 24.9 - fit
Category 3 : employee with bmi over 24.9 - overweight
```

```python
def new_col(new_bmi):
    if (new_bmi<=18.5):
        return "underweight"
    elif (new_bmi>18.5) & (new_bmi<=24.9):
        return "fit"
    else:
        return "overweight"
    
```

```python
df["new_bmi"]=df["bmi"].apply(new_col)
df
```

What is the average expenditure of employees in each category.


```python
df.pivot_table(columns="new_bmi",values="expenditure",aggfunc="mean")
```

Display the bar chart for the same


```python
df.pivot_table(columns="new_bmi",values="expenditure",aggfunc="mean").plot(kind="bar",figsize=(15,8))
plt.title("Average expenditure of employees in each category")
plt.show()
```
